```python
from konlpy.tag import Mecab
from collections import Counter

mec = Mecab()
```

# 형태소 분석+다빈도 단어 추출

```python
# 호텔별 형태소 분석 후 다빈도 단어를 넣을 리스트
reviews = []
for i in tqdm(range(len(copy_df))):
  # 형태소 추출 리스트
  pos_list = []

  # 말뭉치(corpus) 생성
  corpus = "".join(copy_df[hotel_list[i]])
  # corpus에서 형태소 추출
  mec_pos = mec.pos(corpus)

  # 키워드 추출을 위해서는 pos='NNG', 'VA', 'XR', 'MAG'의 빈도를 확인해야 함
  for i in range(len(mec_pos)):
    if mec_pos[i][1] in ('NNG', 'VA', 'XR', 'MAG' ):
      pos_list.append(mec_pos[i][0])

  # 다빈도 단어 추출
  counter = Counter(pos_list)
  # 가장 높은 빈도 단어 50개 출력
  counter.most_common(50)
  # 한글자 명사 제거
  available_counter = Counter({x: counter[x] for x in counter if len(x) > 1 })
  common_noun = available_counter.most_common(50)
  reviews.append(common_noun)
reviews[0]
```

호텔 이름 리스트와 추출된 단어 pairing => dataframe으로 만들기

```python
# dataframe으로 두 리스트 (호텔명 + 다빈도 단어)를 합치기
final_df = pd.DataFrame([x for x in zip(hotel_list, reviews)])
# index명으로 된 칼럼 이름을 새로 지정
mapping = {final_df.columns[0]: 'name', final_df.columns[1]: 'noun_extraction'}
final_df = final_df.rename(columns=mapping)
final_df
```

# LSTM을 위한 형태소 분석(어미)

```python
# 1. 리뷰별 고유번호 부여 (0~)
# 2. 함수 적용
# 3. 공백 컬럼 삭제
# 4. 문장 감성 분석
# 5. 고유번호를 이용해 리뷰별 합치기

def separation(txt):
  mec_pos1 =  mec.pos(txt)
  txt_word_pos = []

  # mecab 결과인 튜플은 수정이 안 되므로 리스트화
  for i in range(len(mec_pos1)):
    txt_word_pos.append([mec_pos1[i][0], mec_pos1[i][1]])

  # 문장 어미에 해당하는 형태소 분류
  txt_word_pos_final = []
  for i in range(len(txt_word_pos)):
    if txt_word_pos[i][1] in ('EF', 'MAJ', 'VV+EF'): # EF(종결 어미), MAJ(MAG : 그런데, 하지만) VV+EFc(됩니다)
      txt_word_pos_final.append(txt_word_pos[i][0])

  # 해당 형태소인 것만 대체어 설정
  for i in range(len(txt_word_pos_final)):
    txt = txt.replace(txt_word_pos_final[i], 'a7ve') # 다른 내용과 겹치면 안되므로 대체어를 특이한 걸로 설정

  # 대체어로 텍스트 분리
  txt_separation = txt.split('a7ve')
  return txt_separation
```

```python
final_result = pd.DataFrame()

# 리뷰별 토큰화 진행 + df화
for i in tqdm(range(len(df_sentiment))):
  if df_sentiment['Code'][i] == i:
    sentence = df_sentiment.iloc[i]
    final_result = final_result.append(pd.Series(sentence, name = df_sentiment['Code'][i])) # 원본 preprocessed_review 삽입
    sentence = df_sentiment['preprocessed_review'][i]
    sentense_separation = separation(sentence) # 고유번호에 따라 preprocessed_review 가져오고 separation 함수 적용

    final_result = final_result.append(pd.Series(sentence, name = df_sentiment['Code'][i])) # 원본 preprocessed_review 삽입

    for k in range(len(sentense_separation)):
      final_result = final_result.append(pd.Series([sentense_separation[k]], name = df_sentiment['Code'][i]))


final_result = final_result.rename(columns={ 0 :'separate_review'})
final_result = final_result.fillna(method = 'pad') # 이전 데이터로 NaN값 채우기
final_result = final_result[['Code', 'name', 'review', 'preprocessed_review', 'sentiment', 'separate_review']]
final_result.head(10)
```

```python
result = final_result.copy()
result = result.dropna()
result = result.reset_index()
result = result.drop(columns='index')
result
```

```python
count = 0
for i in range(len(list(result['separate_review']))):
    if list(result['separate_review'])[i] == '' or list(result['separate_review'])[i] == ' ' or list(result['separate_review'])[i] == '  ':
      count += 1
count
```

```python
# 빈 데이터 drop
droplist = ['', ' ', '  ']
final_token = pd.DataFrame()

for i in tqdm(range(len(result))):
  separate_list = list(result['separate_review'])[i]
  if separate_list != droplist[0] and separate_list != droplist[1] and separate_list != droplist[2]:
    final_token = final_token.append(result.iloc[i])

final_token = final_token.dropna(axis=0)
final_token = final_token[['Code', 'name', 'review', 'preprocessed_review', 'sentiment', 'separate_review']]
final_token
```

```python
# 중복 제거
final_token_df = pd.DataFrame()
for i in tqdm(range(len(list(final_token['separate_review']))-1)):
  if list(final_token['separate_review'])[i] != list(final_token['separate_review'])[i+1]:
    final_token_df = final_token_df.append(final_token.iloc[i])

final_token_df = final_token_df.reset_index()
final_token_df = final_token_df.drop(columns='index')
```
